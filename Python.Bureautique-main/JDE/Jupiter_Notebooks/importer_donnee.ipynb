```python
import pandas as pd

# File path
path = "./Data/eleves.csv"

# Define data types for columns
dtype = {
    'prenom': 'string',
    'nom': 'string',
    'sexe': 'string',
    'classe': 'int'
}

# Read the CSV file with specified data types
df = pd.read_csv(path, encoding='latin1', dtype=dtype, parse_dates=['date_naissance'])
df.info()

#df['date_naissance'] = df['date_naissance'].dt.strftime('%d/%m/%Y')

# Display the dataframe
df.head(50)
df.tail(10)
```

```python
# Sélectionner les colonnes 'nom' et 'prenom'
df[['nom', 'prenom']].head(5)
```

```python
# Filtrer les élèves de la classe 1
df[df['classe'] == 1]
```

```python
# Filter sur les élèves de la classe 1 et de sexe masculin et qui sont née après 2008
filtre = df[(df['classe'] == 3) & (df['sexe'] == 'Garçon') & (df['date_naissance'] > '2008')]
```

```python
# Trier par date de naissance en ordre décroissant
filtre.sort_values('date_naissance', ascending=False)
```

```python
# Ajouter une colonne 'age' calculée à partir de la date de naissance
from datetime import datetime
now = datetime.now()
df['age'] = now.year - df['date_naissance'].dt.year
df.head()
```

```python
# transformation des données
# tri des élèves par age
donnee_triee = df.sort_values('age', ascending=False)
donnee_triee
```

```python
# calcul statistique
# Calcul de l'age moyen
age_moyen = df['age'].mean()
print("L'âge moyen des élèves est de ", age_moyen)
```

```python
# Appliquer une fonction sur une colonne
def sexe_to_int(sexe):
    if sexe == 'Garçon':
        return 1
    else:
        return 0
    
donnee_triee['sexe_int'] = donnee_triee['sexe'].apply(sexe_to_int)

# Modifie l'affichage des colonnes
donnee_triee = donnee_triee[['id', 'prenom', 'nom', 'sexe', 'sexe_int', 'date_naissance', 'classe', 'age']]
donnee_triee.head()
```
