```python
import pandas as pd
```

```python
# récupération du fichier 'utilisateurs.csv' dans le dossier 'Data/Reseaux'
path = r"C:\Users\orsys\Documents\GitHub\python\Python.Bureautique-main\JDE\Jupiter_Notebooks\Data\Reseaux\utilisateurs.csv"
df_utilisateurs = pd.read_csv(path, sep=',', encoding='latin-1')
df_utilisateurs
```

```python
# récupération du fichiers 'reseaux_sociaux.csv' dans le dossier 'Data/Reseaux'
df_donnees = pd.read_csv('Data/Reseaux/reseaux_sociaux.csv', sep=',', encoding='latin-1')
df_donnees
```

```python
# renommer la colonne 'id' du dataframe 'df_utilisateurs' en 'id_utilisateur'
df_utilisateurs.rename(columns={'id': 'id_utilisateur'}, inplace=True)
df_utilisateurs
```

```python
# renommer la colonne 'id' du dataframe 'df_donnees' en 'id_data'
df_donnees.rename(columns={'id': 'id_data'}, inplace=True)
df_donnees
```

```python
# Merge des deux dataframes 'df_utilisateurs' et 'df_donnees' en utilisant la colonne 'adresse IP' comme clé
df_merge= pd.merge(df_utilisateurs, df_donnees, on='adresse IP', how='left')
df_merge
```

```python
# Calculez le nombre d'heures total passé par utilisateur (utilisez l'id utilisateur)
total_hours = df_merge.groupby('id_utilisateur')['temps passé'].sum()
display(total_hours)
df_merge['total_hours'] = df_merge['id_utilisateur'].map(total_hours)
display(df_merge)
```

```python
# Calculez le nombre d'heures total passé par utilisateur (utilisez le nom et prénom de l'utilisateur).
total_hours_by_user = df_merge.groupby(['prénom', 'nom'])['temps passé'].sum()
df_merge['total_hours_2'] = df_merge.apply(lambda x: total_hours_by_user[(x['prénom'], x['nom'])], axis=1)
df_merge

df_tps_passe2 = df_merge.groupby(['prénom', 'nom'])['temps passé'].sum().reset_index()
df_tps_passe2
```

```python
# Calculez le nombre d'heures total passé sur chaque réseau social.
total_hours_by_social_network = df_merge.groupby('site')['temps passé'].sum()
df_merge['total_hours_by_social_network'] = df_merge['site'].map(total_hours_by_social_network)
df_merge
```

```python
# Autre façon de faire l'agrégation
df_merge.groupby('site')['temps passé'].sum()
df_tps_total = df_merge.groupby('id_utilisateur')['temps passé'].sum().reset_index()
df_tps_total
```

```python
# Au dataframe obtenu précédement, ajoutez le pourcentage de temps passé pour chaque réseau social.
df_reseau_social = df_merge.groupby('site')['temps passé'].sum()

# Créer un nouveau dataframe à partir de df_reseau_social
df_new = pd.DataFrame({'site': df_reseau_social.index, 'somme_temps_passé': df_reseau_social.values})

# Afficher le nouveau dataframe
print(df_new.sort_values('somme_temps_passé', ascending=False))

total_time = df_reseau_social.sum()


df_new['percentage'] = round(df_new['somme_temps_passé'] / total_time * 100, 1)
display(df_new.sort_values('percentage', ascending=False))
```

```python
# Visualiser le pourcentage de temps passé sur chaque réseau social en utilisant un diagramme circulaire.
import matplotlib.pyplot as plt

# Create a pie chart
plt.pie(df_new['percentage'], labels=df_new['site'], autopct='%1.1f%%')

# Add a title
plt.title('Percentage of Time Spent on Each Social Network')

# Display the chart
plt.show()
```

```python
# afficher le type des colonnes du dataframe 'df_merge'
print(df_merge.dtypes)
# Convertir la colonne 'date de naissance' en datetime
df_merge['date de naissance'] = pd.to_datetime(df_merge['date de naissance'])

# Afficher le type des colonnes du dataframe 'df_merge'
print(df_merge.dtypes)
df_merge
```

```python
# Au dataframe `df_merge`, ajoutez une colonne `année de naissance` qui contiendra l'année de naissance de chaque utilisateur (au format int).
df_merge['année de naissance'] = df_merge['date de naissance'].dt.year
# Convertir la colonne 'année de naissance' en int
df_merge['année de naissance'] = df_merge['année de naissance'].astype(int)
print(df_merge.dtypes)
df_merge
```

```python
# Créez un tableau croisé dynamique dans lequel on veut afficher le temps passé en moyenne sur chaque réseau social en fonction de l'année de naissance de l'utilisateur.
pivot_table = pd.pivot_table(df_merge, index='site', columns='année de naissance', values='temps passé', aggfunc=lambda x:round(x.mean(), 1))
pivot_table
```

```python
# Afficher la pivot table en graphique 
pivot_table.plot(kind='bar', figsize=(15, 10))
plt.title('Average Time Spent on Each Social Network by Birth Year')
plt.xlabel('Social Network')
plt.ylabel('Average Time Spent')
plt.show()
```
