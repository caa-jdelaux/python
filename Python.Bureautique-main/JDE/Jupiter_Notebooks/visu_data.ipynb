```python
# Visualiser une courbe sin(x) avec matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Donn√©es
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

print(x)
print(y)
```

```python
# Tracer la courbe
plt.plot(x, y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Courbe sin(x)')

plt.legend()
plt.show()
```

```python
# Cr√©er plusieurs courbes
y2 = np.cos(x)
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Courbe sin(x) et cos(x)')
plt.legend()
plt.show()
```

```python
# Afficher l'avancement des tests OK / KO / non test√© dans le temps
import matplotlib.pyplot as plt
import pandas as pd

data = {'Timestamp': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04'],
    'OK': [1, 5, 9, 10],
    'KO': [0, 1, 1, 0],
    'Not Tested': [9, 4, 0, 0]}

df = pd.DataFrame(data)
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Timestamp'] = df['Timestamp'].dt.strftime('%d/%m/%y')

# Afficher les donn√©es
print(df)



# Cr√©er un histogramme
ax = df.set_index('Timestamp').plot(kind='bar', stacked=True, color=['green', 'red', 'blue'], width=1, edgecolor='none')

# V√©rifier si la valeur est diff√©rente de 0 et afficher la l√©gende correspondante
for col in df.columns[1:]:
    for i, val in enumerate(df[col]):
        if val != 0:
            ax.text(i, val/2, str(val), va='bottom', ha='center')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_xlabel('date')
plt.show()
```

```python
# Visualiser des coordonn√©es g√©ographiques sur une carte en utilisant folium
import folium

# D√©finir les coordonn√©es du centre de la carte
latitude = 48.8566
longitude = 2.3522
coordinates = [ {'name': 'Paris', 'latitude': 48.8566, 'longitude': 2.3522},
                {'name': 'New York', 'latitude': 40.7128, 'longitude': -74.0060},
                {'name': 'Tokyo', 'latitude': 35.6895, 'longitude': 139.6917}]


# Create a map object
m = folium.Map(location=[latitude, longitude], zoom_start=12)

# Add markers for each coordinate
for coord in coordinates:
    folium.Marker(location=[coord['latitude'], coord['longitude']], popup=coord['name']).add_to(m)

# Display the map
m
```

```python
# Graphique int√©ractif avec panel
import panel as pn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Cr√©er des donn√©es al√©atoires
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.random.randn(100)*np.sin(x)

# Cr√©er un DataFrame √† partir des donn√©es
df = pd.DataFrame({'x': x, 'y': y})

# Cr√©er une figure matplotlib
fig, ax = plt.subplots()
ax.plot(df['x'], df['y'])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Graphique interactif')

# Cr√©er un panneau Panel pour afficher la figure
panel = pn.pane.Matplotlib(fig)

# Afficher le panneau
panel.show()
```
