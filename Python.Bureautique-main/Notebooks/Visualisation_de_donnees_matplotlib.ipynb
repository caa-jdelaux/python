```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Données
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Tracer la courbe
plt.plot(x, y, label='sin(x)')
plt.title('Courbe sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.legend()
plt.show()
```

```python
x = np.arange(0, 10, 0.1)

# Plusieurs courbes sur le même graphique
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='#FF0000')
plt.title('Courbes sin(x) et cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, which='both')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.legend()
plt.show()
```

```python
# Tracer des histogrammes avec Matplotlib
# Données
donnees = np.random.randn(1000)

# Tracer l'histogramme
plt.hist(donnees, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogramme des données aléatoires')
plt.xlabel('Valeurs')
plt.ylabel('Fréquence')
plt.show()
```

```python
# Données pour le Pie Chart
labels = ['Catégorie 1', 'Catégorie 2', 'Catégorie 3']
sizes = [30, 45, 25]

# Tracer le Pie Chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Exemple de Pie Chart')
plt.show()
```
